### Posterior

$w_{i} = \frac{Arctan(\pi v_{i})}{\pi} + \frac{1}{2}$ Where $v_{i}$ ~ $Gauss(0,\Sigma_{a,b})$

```python
def logGaussian(inverseSigma, mean, x):
    v = np.array([(1/pi)*(tan((pi)*(y-1/2))) for y in x])
    return (-1/2)*(v-mean).dot(inverseSigma).dot((v-mean).T)
```

```python
## Compute the lambdas
def strength(i,j,p,vec):
    summ = 0
    for l in range(p):
        summ += vec[i,l]*vec[j,l]
    return summ

# n : Population / p : Number of communities / k : Number of ranks / a,b : Gamma parameters / w = Community parameter
#/ rho : Data rho[i,j] jth selection of the ith person
def posterior(n,p,k,inverseSigma,mean ,w,rho):
    vec = w.reshape((n,p))
    summa = 0
    for i in range(n):
        summa += logGaussian(inverseSigma,mean,vec[i])
        totalStrength = 0
        listIndex = [l for l in range(n)]
        del listIndex[i]
        for u in (listIndex):
            totalStrength += strength(i,u,p,vec)
        stateStrength = 0
        for j in range(k):
            totalStrength -= stateStrength
            summa -= log(abs(totalStrength) + 0.000000000000001)
            stateStrength = strength(i,rho[i,j],p,vec)
            summa += log(abs(stateStrength) + 0.000000000000001)
    return summa
```

```python
def optimize(n,p ,k,inverseSigma,mean,rho,w0):
    costFunction = lambda w : (-1)*posterior(n,p,k,inverseSigma,mean,w,rho)
    return opt.minimize(costFunction,w0,bounds = [(0,1) for i in range(n*p)])
```

### Data Generation

```python
## Convert the array from the Sampson's DataSet in the format of the data in the dissertation Proposal
def dataProcessing(arr,k):
    r,s = arr.shape
    sol = np.zeros((r,k),dtype = 'int')
    for i in range(r):
        for j in range(1,k+1):
            for l in range(s):
                if arr[i,l] == j:
                    sol[i,j-1] = int(l)
    return sol
```

```python
(listData,listLabel) = generateDataSet()
arr = listData[0]
n,p = arr.shape
k = 3
p = 3
a = 1
b = 1
rho = dataProcessing(arr,k)
```

```python
E = np.array([[2.5,-1,-1],[-1,2.5,-1],[-1,-1,2.5]])
sigmaInverse = np.linalg.inv(E)
mean = np.array([0,0,0])

w0 =  np.random.rand(n*p)
result = optimize(n,p ,k,sigmaInverse,mean,rho,w0)
minimum, xminimum = result.fun, result.x
for i in range(200):
    w0 =  np.random.rand(n*p)
    result = optimize(n,p ,k,sigmaInverse,mean,rho,w0)
    if result.fun < minimum:
        minimum, xminimum = result.fun, result.x
        print(minimum)
print(minimum, xminimum)
```

```python
x = xminimum.reshape((18,3))
```

### Plotting graphs

```python
g = Graph(directed=True)
g.add_vertices(n)
for i in range(n):
    for j in range(k):
        g.add_edge(i,rho[i,j], width=j+1)
g.vs["label"] = listLabel
colors = ["red", "green", "blue"]
g.vs["color"] = [colors[np.argmax(x[i])] for i in range(n-1)]
plot(g)
```
