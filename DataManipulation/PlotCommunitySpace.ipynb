## Community Plot

#### The aim of this notebook is to produce plot that give intuition behind community model

```python
import scipy.optimize as opt
import numpy as np
from math import *
import matplotlib.pyplot as plt
from DataSet import generateDataSet
from igraph import *
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.style
import matplotlib as mpl
mpl.style.use('classic')
from igraph import *
from DataSet import generateDataSet
import seaborn as sns; sns.set()
import scipy.stats as gam
```

```python
ax = 2
thetax = 2
ay = 2
thetay = 4
lAxes = np.linspace(0,10,250)
xList = np.array([[gam.gamma.pdf(x, ax,0,thetax) for x in lAxes]])
yList = np.array([[gam.gamma.pdf(x, ay,0,thetay) for x in lAxes]])
mapDistrib = (yList.T).dot(xList)
ax = sns.heatmap(mapDistrib,xticklabels=False, yticklabels=False)
plooo = ax.get_figure()
plooo.savefig("heatMap.png")
```

```python
lAxes = np.linspace(0,10,250)
mapDistrib = np.array([[5*i+0*j for i in lAxes] for j in lAxes])
ax = sns.heatmap(mapDistrib,xticklabels=False, yticklabels=False)
plooo = ax.get_figure()
plooo.savefig("heatMap.png")
```
